Разбейте эти датафреймы на тренировочные (X_train, y_train) и тестовые (X_test, y_test) с помощью функции train_test_split так, чтобы размер тестовой выборки

Создайте модель линейной регрессии под названием lr с помощью класса LinearRegression из модуля sklearn.linear_model.
Обучите модель на тренировочных данных (используйте все признаки) и сделайте предсказание на тестовых.

Boston House Prices dataset был удален из scikit-learn начиная с версии 1.2. Для выполнения домашнего задания использован California housing dataset из встроенных наборов данных библиотеки sklearn.

```python
# Импортируем необходимые библиотеки
import pandas as pd
import numpy as np
from sklearn.datasets import fetch_california_housing
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error

# Загрузим данные
housing = fetch_california_housing()
X = pd.DataFrame(housing.data, columns=housing.feature_names)
y = pd.Series(housing.target)

# Разделим данные на тренировочные и тестовые выборки
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Создадим модель линейной регрессии
lr = LinearRegression()

# Обучим модель на тренировочных данных
lr.fit(X_train, y_train)

# Сделаем предсказание на тестовых данных
y_pred = lr.predict(X_test)

# Выведем результаты
print("Mean Squared Error:", mean_squared_error(y_test, y_pred))
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import GridSearchCV, train_test_split
from sklearn.metrics import roc_auc_score

# Настройка для отображения всех столбцов датафрейма
pd.options.display.max_columns = 100
```

```python
# Загрузка датасета
df = pd.read_csv('creditcard.csv')

# Проверка на несбалансированность выборки
print(df['Class'].value_counts(normalize=True))

# Проверка на наличие пропусков и типы данных
print(df.info())

# Просмотр первых 10 строк датафрейма
print(df.head(10))
```

```python
# Создание датафрейма X из датафрейма df, исключив столбец Class
X = df.drop(columns='Class')

# Создание объекта Series y из столбца Class
y = df['Class']

# Разделение данных на тренировочные и тестовые наборы
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=100, stratify=y)

# Просмотр информации о форме данных
print("Shape of X_train:", X_train.shape)
print("Shape of X_test:", X_test.shape)
print("Shape of y_train:", y_train.shape)
print("Shape of y_test:", y_test.shape)
```

```python
# Задание параметров для поиска по сетке
parameters = [{'n_estimators': [10, 15],
               'max_features': np.arange(3, 5),
               'max_depth': np.arange(4, 7)}]

# Создание модели GridSearchCV
grid_search = GridSearchCV(estimator=RandomForestClassifier(random_state=100),
                           param_grid=parameters,
                           scoring='roc_auc',
                           cv=3)

# Обучение модели на тренировочном наборе данных
grid_search.fit(X_train, y_train)

# Просмотр параметров лучшей модели
best_params = grid_search.best_params_
print("Best parameters:", best_params)
```

```python
# Предсказание вероятностей классов
y_pred_proba = grid_search.predict_proba(X_test)[:, 1]

# Вычисление AUC на тестовых данных
auc = roc_auc_score(y_test, y_pred_proba)
print("AUC on test data:", auc)

# Дополнительно: Вычисление AUC на тренировочных данных для сравнения
y_train_pred_proba = grid_search.predict_proba(X_train)[:, 1]
auc_train = roc_auc_score(y_train, y_train_pred_proba)
print("AUC on training data:", auc_train)
```

```python
from sklearn.datasets import load_wine

# Шаг 1: Загрузка датасета Wine
data = load_wine()

# Шаг 2: Проверка типа данных и ключей в структуре данных
print("Type of data:", type(data))  # Это будет <class 'sklearn.utils.Bunch'>
data_keys = list(data.keys())
print("Keys of the dataset:", data_keys)

# Шаг 3: Просмотр данных, описания и названий признаков
# Данные
X_data = data.data
print("Data shape:", X_data.shape)

# Описание
description = data.DESCR
print("Description of the dataset:\n", description)

# Названия признаков
feature_names = data.feature_names
print("Feature names:", feature_names)

# Шаг 4: Количество классов и их названия
n_classes = len(data.target_names)
print("Number of classes:", n_classes)

class_names = data.target_names
print("Class names:", class_names)

# Шаг 5: Создание датафрейма X из данных и названий признаков
X = pd.DataFrame(X_data, columns=feature_names)
print("First 5 rows of DataFrame X:")
print(X.head())

# Шаг 6: Проверка размера датафрейма и наличие пропущенных значений
print("Shape of DataFrame X:", X.shape)
missing_values = X.isnull().sum().sum()
print("Total missing values in DataFrame X:", missing_values)

# Шаг 7: Добавление поля 'target' в датафрейм X
X['target'] = data.target.astype(np.int64)
print("First 5 rows of DataFrame X with target:")
print(X.head())

# Шаг 8: Построение матрицы корреляций
X_corr = X.corr()
print("Correlation matrix X_corr:")
print(X_corr)

# Шаг 9: Создание списка признаков с высокой корреляцией с целевой переменной
high_corr = [col for col in X_corr.columns if col != 'target' and abs(X_corr.loc['target', col]) > 0.5]
print("High correlation features with target:", high_corr)

# Шаг 10: Модификация датафрейма X
# Удаление поля 'target'
X = X.drop(columns='target')

# Добавление полей с квадратами значений признаков из списка high_corr
for feature in high_corr:
    X[feature + '_2'] = X[feature] ** 2

# Вывод описания полей датафрейма X
print("Description of the modified DataFrame X:")
print(X.describe())
```
